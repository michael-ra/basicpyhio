## Data Preprocessing

```python
# load the dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/MagMueller/BasicPython/main/pima-indians-diabetes.csv', 
                      delimiter=',', 
                      names=["Number of pregnancies", "Plasma glucose", "mm Hg blood press.", 
                             "Triceps skin thickness", "2 hour serum insulin mu U per ml", 
                             "BMI", "pedigree diabetes function", "Age","has diabetes"])
```

After loading the data we convert it to a pandas dataframe which will facilitate further processing. After that we want to get a first overview over the data.

```python
#df = pd.DataFrame.from_records(dataset)
df = dataset.copy()
print("length of dataset: {}".format(len(df)))
df.head()
```

We now know that our dataframe has 768 rows and 9 columns. As a next step we check whether the data comprises any NAs.

```python
df.isna().sum()
```

Since there are no NAs in the dataset we can move on to get a more detailed overview over our data by using pandas describe function. The produced dataframe is then transposed to improve readability.

```python
df.describe().transpose()
```

We can see that the min of some measurements is 0 where it is clearly only a replacement for NAs (eg. Blood pressure, BMI..). So we replace these missing measurements with a better substitution, which will  namely the median of the respective variable 

```python

# We exclude the first and the last column since 0 is a plausible value for them
for colname in df.columns[1:-1]: 
    mask = df[colname] == 0.00 #use a mask to prevent working with a copy
    df.loc[mask, colname] = df[colname].median()

df.describe().transpose()
```

We can see that for Serum insulin the values range from 14 to 846 while for the diabetes pedigree function the values values stay between 0.078 and 2.42. This large disparity of value size could yield problems in the learning process, therefore we normalized the data by dividing every item of a column by the maximum value of the column.

```python
#normalize by deviding every element of the dataframe by the maximum value of the dataframe
df_unscaled = df.copy()
for column in df.columns:
    df[column] = df[column]  / df[column].abs().max()


df.describe().transpose()
```

# **Data preparation and splits**

## Split data

```python
# format: (Model,Accuracy,Parameters)^n
model_acc = []
features = (dataset.iloc[:,:8]).values
target = (dataset.loc[:,'result']).values

# dataset sizes
testSize = 0.3 #%
trainSize = 0.7 #%
validSize = 0.35 #%

# train test split https://stackoverflow.com/a/51006816/16329933
x_train, x, y_train, y = sk.model_selection.train_test_split(features,target,train_size = trainSize)
x_val, x_test, y_val, y_test = sk.model_selection.train_test_split(x,y,train_size=validSize)

print("Sizes of")
print("x_train:    ",len(x_train))
print("y_train:    ",len(y_train))
print("x_val:      ",len(x_val))
print("y_val:      ",len(y_val))
print("x_test:     ",len(x_test))
print("y_test:     ",len(y_test))
```

Prepare the SVM data using a RobustScaler with modified quantiles.

```python
# modified RobustScaler, using 30,70 quantile ranges
svm_scaler = sk.preprocessing.RobustScaler(with_centering=True,with_scaling=True,quantile_range=(30.0, 70.0),copy=True)
x_train_svm = svm_scaler.fit_transform(x_train)
x_val_svm  = svm_scaler.fit_transform(x_val)
x_test_svm = svm_scaler.fit_transform(x_test)
```

```python
# normalize the data for the FFANN
x_train_nn = sk.preprocessing.normalize(x_train)
x_val_nn = sk.preprocessing.normalize(x_val)
x_text_nn = sk.preprocessing.normalize(x_test)
```

## SVM

```python
support_vector_classifier = sk.svm.SVC(kernel='rbf',class_weight = 'balanced',random_state = 1)

# train the svm
support_vector_classifier.fit(x_train_svm,y_train)

# validate the classifier
y_pred = support_vector_classifier.predict(x_val_svm)

# print the accuracy and store it
print("__ SVM METRICS __")
# calculate the accuracy
met_acc = sk.metrics.accuracy_score(y_val, y_pred)
print("--------------------")
print("Accuracy:",met_acc)
# append accuracy to global metrics
model_acc.append(("SVM",met_acc,"none"))
print("--------------------")
print("Metric table:")
print(sk.metrics.classification_report(y_val, y_pred, digits=3, output_dict=False))
```

## KNN
Using the svm data for knn too, as it requires scaling, not normalization.

```python
val_scores = []
k_values = []
```

```python
# set neighbors N for knN
neighbords = 1
# lets go until 30 n to factor into the algorithm
while(neighbords < 30):
  i = neighbords
  # init the knn classifier with i as neighbors
  knn = sk.neighbors.KNeighborsClassifier(i)
  # fit the model from above
  knn.fit(x_train_svm,y_train)

  knn_y_pred = knn.predict(x_val_svm)
  # get the accuracy
  met_acc = sk.metrics.accuracy_score(y_val, knn_y_pred)
    
  # append the current model accuracy to the list
  val_scores.append(met_acc)
  k_values.append(i)
  neighbords = neighbords+1

# get the best performing models accuracy and parameter
print("__ KNN METRICS __")
print("--------------------")
best_acc = max(val_scores)
print("Best Accuracy:",best_acc)
print("Accuracy best at n =",val_scores.index(best_acc))
# append model parameters and accuracy to global list
model_acc.append(("KNN",met_acc,val_scores.index(best_acc)))
print("--------------------")
print("Metric table:")
print(sk.metrics.classification_report(y_val, knn_y_pred, digits=3, output_dict=False))
```

## Define the ANN model

```python
# PReLU
from keras.layers.advanced_activations import PReLU
```

In the following code we experiemented with some regularizers, dropout and a layer structure. For the basic model, which is suggested in this case as we have limited data avaliable (round 500 rows, which is too few to apply fancy Dropout, regularization and find the best minima), we use a kinda-deep FeedForwardNeuralNetwork. 

```python
model = Sequential()
model.add(Dense(8, input_dim=8, activation='relu')) 
model.add(Dense(16, activation='relu'))
model.add(Dense(24, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
